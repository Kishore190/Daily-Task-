```python
import pandas as pd
df=pd.read_csv("C:\\Users\\XNovaQ.8\\Downloads\\spam.csv", encoding='latin-1')
df.head()
```

```python
df = df.iloc[:, [0, 1]] 
df.columns = ['label', 'message']
df.head()
```

```python
df['label'] = df['label'].map({"ham":0, "spam":1})
df.head()
```

```python
df['length'] = df['message'].apply(len)
df
```

```python
df['message']=df['message'].str.lower()
df
```

```python
import re
def clean_text(text):
    text = str(text).lower()                   
    text = re.sub(r'[^a-z\s]', '', text)       
    text = re.sub(r'\s+', ' ', text).strip()   
    return text

df['message'] = df['message'].apply(clean_text)
df
```
