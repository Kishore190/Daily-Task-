```python
import pandas as pd

file_name = "C:\\Users\\XNovaQ.8\\Downloads\\archive (1)\\student\\student-mat.csv"
df = pd.read_csv(file_name, sep=';')

print("--- Initial Data Info ---")
df.info()
print("\n--- Missing Values Check ---")
print(df.isnull().sum().to_string())
print("\n--- Duplicate Rows Check ---")
print(f"Number of duplicate rows: {df.duplicated().sum()}")
```

```python

binary_map = {'yes': 1, 'no': 0}
binary_cols = ['schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic']
for col in binary_cols:
    df[col] = df[col].map(binary_map)

df['sex'] = df['sex'].map({'F': 0, 'M': 1}) 
df['school'] = df['school'].map({'GP': 0, 'MS': 1})
df['address'] = df['address'].map({'U': 1, 'R': 0}) 
df['famsize'] = df['famsize'].map({'LE3': 0, 'GT3': 1}) 
df['Pstatus'] = df['Pstatus'].map({'T': 1, 'A': 0}) 
df.head()
```

```python

object_cols = df.select_dtypes(include=['object']).columns
print(f"\nRemaining object columns to One-Hot Encode: {list(object_cols)}")

df = pd.get_dummies(df, columns=object_cols, drop_first=True, dtype=int)
```

```python

print("\n--- DataFrame Head after Encoding ---")
print(df.head())
print("\n--- Final Data Info ---")
df.info()

df.to_csv('student-mat_preprocessed.csv', index=False)
print("\nPreprocessed data saved to 'student-mat_preprocessed.csv'")
```
